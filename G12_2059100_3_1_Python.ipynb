PROGRAMA DE CASO

Se organiza un evento de entusiastas de la música en un bar de la ciudad, donde habrá música en vivo, y la entrada es a cambio de una cooperación voluntaria con fines de caridad. Para entrar, todas las personas deben tener un boleto, que está numerado, y que va del 100 al 999.

Cada persona, al entrar, registra su número de boleto, posteriormente, debe registrar su correo electrónico, su nombre, sus apellidos, clave de sexo (M=Masculino, F=Femenino, N=No binario), y la aportación voluntaria que desea hacer, misma que paga en ese momento.

Todos los textos deben estar convertidos a mayúsculas. No se programan validaciones: los datos se capturarán siguiendo Happy Path (escenario de funcionamiento ideal, donde vamos capturando lo que sabemos que funciona sin problemas).

El programa captura los datos en un ciclo continuo, del cuál solo se puede salir omitiendo el número de boleto cuando se solicita. Cuando se registran los datos de un participante, el programa nos dirá “Gracias, contigo ya somos {asistentes} personas, y hemos recaudado ${monto_acumulado} hasta el momento”. Donde asistentes será un contador de las personas que se han registrado, mientras que monto_acumulado es un acumulador que va sumando todas las aportaciones voluntarias que se vayan haciendo. El programa solo debe validar que los datos sean del tipo correcto, usando Try Catch.

Cuando en la captura del número de boleto se omita el número de boleto, el programa deberá preguntar si ya quieres abandonar el programa. Si contestas que NO, seguirá preguntando datos. Si le dices que SI, deberá mostrar una tabla donde se muestren todos los datos capturados, si le das algo diferente de SI o NO, debe decirte que no se reconoce la acción, y que intente de nuevo.

Además, deberá mostrarse qué porcentaje de personas se identificaron como del sexo masculino (M), femenino (F), o no binario (N).

Los datos deben guardarse en un diccionario, donde la llave es el número de boleto, mientras que el resto de los datos deberán guardarse primero en una lista, misma que se guardará como valor de la entrada del diccionario.

No se debe permitir la captura de un mismo número de boleto dos veces. Si se intenta, el programa deberá decir “Error de captura. Ese boleto ya está capturado. Intenta con otro.”, y vuelve a capturar.

| Dato | Tipo de Dato | Origen | Variable | Fórmula |
|----|----|----|----|----|
| Número de Boleto | Int  | Se pregunta  | numero_boleto |  |
| Correo Electrónico | String | Se pregunta | correo |   |
| Nombre | String | Se pregunta  | nombre |   |
| Apellidos | String | Se pregunta | apellidos  |  |
| Sexo  | String | Se pregunta | sexo |  |
| Aportación | Float  | Se pregunta | aportacion |  |
| Total de Asistentes | Int | Se calcula | asistentes |  |
| Monto Recaudado | Float | Se calcula | monto_acumulado  |  |
| % Participantes Masculinos | Float | Se calcula | porcentaje_masculino | (porcentaje_masculino / asistentes) * 100  |
| % Participantes Femeninos | Float | Se calcula | porcentaje_femenino | (porcentaje_femenino / asistentes) * 100  |
| % Participantes No Binarios | Float | Se calcula  | porcentaje_no_binario | (porcentaje_no_binario / asistentes) * 100  |






Diagrama de Flujo.png

```python
# Diccionario para almacenar los datos
registros = {}

# Variables
asistentes = 0
monto_acumulado = 0
porcentaje_masculino = 0
porcentaje_femenino = 0
porcentaje_no_binario = 0

while True:
    try:
        # Captura del número de boleto
        numero_boleto = input("Ingresa el numero de tu boleto 100-999: ")

        # Salir del programa si no se ingresa un número de boleto
        if numero_boleto == "":
            salir = input("¿Quieres salir del programa? (SI/NO): ").upper()
            if salir == "SI":
                # Mostras la tabla de datos
                print("\n--- Tabla de datos ---")
                print("Número de Boleto | Correo Electrónico | Nombre | Apellidos | Sexo | Aportación")
                for boleto, datos in registros.items():
                    print(f"{boleto}              | {datos[0]} | {datos[1]} | {datos[2]} | {datos[3]} | ${datos[4]}")
                print("\nEstadísticas:")
                print(f"Total de asistentes: {asistentes}")
                print(f"Monto recaudado: ${monto_acumulado}")
                print(f"Porcentaje de participantes masculinos: {porcentaje_masculino / asistentes * 100}%")
                print(f"Porcentaje de participantes femeninos: {porcentaje_femenino / asistentes * 100}%")
                print(f"Porcentaje de participantes no binarios: {porcentaje_no_binario / asistentes * 100}%")
                break
            elif salir == "NO": # Si inserta "NO" comienza el ciclo de nuevo
                continue
            else:
                print("Acción no reconocida. Intenta de nuevo.")
                continue

        # Comprobar que el numero no este repetido
        if numero_boleto in registros:
            print("Error de captura. Ese boleto ya está capturado. Intenta con otro.")
            continue

        numero_boleto = int(numero_boleto) # Convertir el numero a entero


        correo = input("Ingresa tu correo electrónico: ").upper()         # Captura de correo en mayusculas
        nombre = input("Ingresa tu nombre: ").upper()                     # Captura de nombre en mayusculas
        apellidos = input("Ingresa tus apellidos: ").upper()              # Captura de apellidos en mayusculas
        sexo = input("Ingresa tu clave de sexo (M/F/N): ").upper()        # Captura de las siglas del sexo en mayusculas
        aportacion = float(input("Ingresa tu aportación voluntaria: "))   # Captura de la aportación

        # Porcentajes de sexo
        if sexo=="M":
            porcentaje_masculino+= 1
        elif sexo=="F":
            porcentaje_femenino+= 1
        elif sexo=="N":
            porcentaje_no_binario+= 1

        # Contador y acumulador
        asistentes+= 1
        monto_acumulado+= aportacion

        # Almacenar datos en el diccionario
        registros[numero_boleto] = [correo, nombre, apellidos, sexo, aportacion]

        print(f"Gracias, contigo ya somos {asistentes} personas, y hemos recaudado ${monto_acumulado} hasta el momento.") # Mensaje de agradecimiento

    except:
        print("Error de captura. Por favor, ingresa un número de boleto válido.")

```
